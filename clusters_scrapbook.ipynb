```python
global buff,cl,cn,blen,b,nc,meanArray
cl = [None]*100
cn = 0
blen = 0
b = 0
nc = 0
meanArray = []


for i in range(len(TimepixAData)): # for each buffer
    
    buff = []                              # start a new buffer array
    for k in range(len(TimepixAData[i])): # for each hit in the current buffer
        buff.append(TimepixAData[i][k])   # put the hit in our buffer array
    buff.sort(key = sortThird)              # sort the buffer by ToA
    blen = len(buff)
    
    while(b<blen):         # while still in the buffer
        cn = 0             # start with a fresh cluster
        cl[cn] = b         # consider the first point as part of our first cluster
        cn+=1              # increment the number of hits for this cluster
        c = b+1
        while(c<b+99 and c<blen):                       # as long as we don't have too many hits and still in buffer
            if(abs(buff[c]['ToA']- buff[b]['ToA'] < 100)):     # if hits within a 100 nano seconds
                cl[cn] = c                                     # then count this as part of the cluster and store it
                cn+=1                                          # increment the number of hits for this cluster
            else:
                process_cluster(i)                 # process the cluster
                b+=cn                             # move past this cluster             
                cn = 0                            # restart the counter for the new cluster
                cl = [None]*100                   # restart the cluster array for the new cluster
                nc+=1                             # increment the number of clusters found

                if(c<b+99 and c<blen):            # so long as we're still in the while loop
                    cl[cn] = b                    # consider the current position the first hit in new cluster
                    cn+=1                         # increment the number of hits for this cluster

            c+=1                       # check the next cluster position
        b+=1                           # move on to the next buffer position
        
    b=0
    
#print(meanArray)
    
ROOT.enableJSVis()

c = ROOT.TCanvas("c","Clusters A", 1000,1000)
c.Divide(2,1)

tdposA = Hist2D(256,0,256,256,0,256)
tdposA.GetXaxis().SetTitle("mean x position of the cluster")
tdposA.GetYaxis().SetTitle("mean y position of the cluster")

multA = Hist(50,0,50)
multA.GetXaxis().SetTitle("number of hits in the cluster")
multA.GetYaxis().SetTitle("number of clusters")

c.cd(1)
tdposA.Draw()
c.cd(2)
multA.Draw()
c.Draw()
```
